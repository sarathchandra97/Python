````python
# Simplest plot
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
````

````python
## Sample each plot 
x = np.linspace(0, 2, 100)

# set the values of the x and y
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# set the s labell
plt.xlabel('x label')
plt.ylabel('y label')

# set the title 
plt.title("Simple Plot")

# set that i want legend
plt.legend()

# plot every thing below
plt.show()
````
