```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn import svm
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import Birch
```

```python
x = np.linspace(start=-10, stop=10, num=100)
y = stats.norm.pdf(x,loc=-5, scale=1.5)
tx = np.array(x)
ty = np.array(y)
combined = np.vstack((tx, ty)).T
plt.plot(combined[:,0], combined[:,1])
```

```python
x = np.linspace(start=-10, stop=10, num=100)
y = stats.norm.pdf(x,loc=5, scale=2.5)
tx = np.array(x)
ty = np.array(y)
combined2 = np.vstack((tx, ty)).T
plt.plot(combined2[:,0], combined2[:,1])
```

```python
final_combined = np.concatenate((combined, combined2))
final_combined
```

```python
np.linspace(start=0, stop=0, num=100)
```

```python
y = np.concatenate(np.linspace(start=0, stop=0, num=100),np.linspace(start=0, stop=0, num=100))
```

```python
plt.plot(final_combined[:,0], final_combined[:,1])
```

```python
GMM = GaussianMixture(2)
labels = GMM.fit_predict(final_combined)
labels
```

```python
y1 = np.linspace(start=0, stop=0, num=100)
y2 = np.linspace(start=1, stop=1, num=100)
y12 = np.concatenate((y1,y2))
plt.scatter(final_combined[:,0], final_combined[:,1],c=labels,s=30)
```

```python
clf = svm.SVC(kernel='linear', C=1)
clf.fit(final_combined,y12)
```

```python
y_final = clf.predict(final_combined)
```

```python
plt.scatter(final_combined[:,0], final_combined[:,1],c=y_final,s=30)
```

```python
clf = KMeans(2)
clf.fit(final_combined,y12)
y_kmeans  = clf.predict(final_combined)
plt.scatter(final_combined[:,0], final_combined[:,1],c=y_kmeans,s=30)
```

```python
clf = DBSCAN(2)
y_dbscan  = clf.fit_predict(final_combined)
plt.scatter(final_combined[:,0], final_combined[:,1],c=y_dbscan,s=30)
```

```python
clf = Birch()
y_brich  = clf.fit_predict(final_combined)
plt.scatter(final_combined[:,0], final_combined[:,1],c=y_brich,s=30)
```
